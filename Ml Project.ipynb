```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```python
APP_FOLDER = "E:\sem 3\Machine Learning\Project\PokemonData"

totalFiles = 0
totalDir = 0

dir=[]
fileses=[]
indexed=[]

for base, dirs, files in os.walk(APP_FOLDER):
    #print('Searching in : ',base)
    for directories in dirs:
        totalDir += 1
        
        dir.append(directories)
        

    for Files in files:
        totalFiles += 1
    
    fileses.append(totalFiles)
   
    totalFiles=0
 
del fileses[0]
```

```python
df=pd.DataFrame({'Pokemons': dir , 'Images Count': fileses})


print(df)
```

```python
dfs=df.sample(10)
plt.figure(figsize=(10,8))
plt.plot(dfs['Pokemons'],dfs['Images Count'])
plt.xlabel("Pokemons")
plt.ylabel('Image Count')
plt
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot("Pokemons",'Images Count',data=dfs)
```
